1.  Create a line plot of the provided dataset with years on the x-axis and population on the y-axis.

```python
%pip install --upgrade pip
%pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt


# Sample dataset for years and population
data = {'Year': [2015, 2016, 2017, 2018, 2019, 2020],
        'Population': [7.3, 7.4, 7.5, 7.6, 7.7, 7.8]}  # In billions

df = pd.DataFrame(data)

# Plotting the line chart
plt.plot(df['Year'], df['Population'])
plt.title('World Population Over Time')
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.show()
```

3. Create a bar chart of the provided dataset with car brands on the x-axis and average price on the y-axis.

```python
# Sample dataset for car prices by brand
car_data = {'Brand': ['Audi', 'BMW', 'Mercedes', 'Toyota', 'Honda'],
            'Price': [30000, 50000, 40000, 20000, 18000]}

df = pd.DataFrame(car_data)

# Bar chart
plt.bar(df['Brand'], df['Price'])
plt.title('Car Brands vs. Average Price')
plt.xlabel('Brand')
plt.ylabel('Average Price ($)')
plt.show()
```

4. Create a histogram of the provided dataset with salaries on the x-axis and frequency on the y-axis.

```python
import numpy as np

# Sample dataset for salaries
salaries = np.random.normal(50000, 15000, 1000)  # Mean=50k, SD=15k

# Histogram
plt.hist(salaries, bins=30)
plt.title('Salary Distribution')
plt.xlabel('Salary ($)')
plt.ylabel('Frequency')
plt.show()
```

5. Create a pie chart of the provided dataset with the percentage of car sales of each brand.

```python
# Sample dataset for car sales by brand
car_sales = {'Brand': ['Audi', 'BMW', 'Mercedes', 'Toyota', 'Honda'],
             'Sales': [50000, 60000, 75000, 80000, 90000]}

df = pd.DataFrame(car_sales)

# Pie chart
plt.pie(df['Sales'], labels=df['Brand'], autopct='%1.1f%%')
plt.title('Percentage of Car Sales by Brand')
plt.show()
```

6. Create a stacked bar chart of the provided dataset with the years on the x-axis and the total number of medals won by each country on the y-axis. The stacked bars should show the number of gold, silver, and bronze medals won.

```python
# Sample dataset for medal counts by country (Correcting the data)
medals_data = {'Year': [2015, 2016, 2017, 2018, 2019, 2020],
               'Gold': [46, 38, 29, 24, 18, 12],
               'Silver': [37, 32, 17, 24, 21, 15],
               'Bronze': [38, 18, 17, 27, 23, 19]}

df = pd.DataFrame(medals_data)

# Stacked bar chart
plt.bar(df['Year'], df['Gold'], label='Gold', color='gold')
plt.bar(df['Year'], df['Silver'], bottom=df['Gold'], label='Silver', color='silver')
plt.bar(df['Year'], df['Bronze'], bottom=df['Gold']+df['Silver'], label='Bronze', color='brown')
plt.title('Medals Won Over Years by Country')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.legend()
plt.show()
```

7. Create a box plot of the provided dataset with car prices grouped by brand.

```python
# Sample dataset for car prices by brand
car_data = {'Brand': ['Audi', 'BMW', 'Mercedes', 'Toyota', 'Honda'],
            'Price': [30000, 50000, 40000, 20000, 18000]}

df = pd.DataFrame(car_data)

# Box plot
df.boxplot(column='Price', by='Brand')
plt.title('Box Plot of Car Prices by Brand')
plt.suptitle('')
plt.show()
```

8. Create a violin plot of the provided dataset with car prices grouped by brand.

```python
import seaborn as sns

# Violin plot using the same car price data
sns.violinplot(x='Brand', y='Price', data=df)
plt.title('Violin Plot of Car Prices by Brand')
plt.show()
```

9. Create a hexbin plot of the provided dataset with car prices on the x-axis and mileage on the y-axis.

```python
# Sample dataset for car prices and mileage
car_data = {'Price': [30000, 50000, 40000, 20000, 18000],
            'Mileage': [20, 18, 22, 25, 30]}

df = pd.DataFrame(car_data)

# Hexbin plot
plt.hexbin(df['Price'], df['Mileage'], gridsize=20, cmap='Blues')
plt.colorbar(label='Frequency')
plt.title('Hexbin Plot of Car Prices vs. Mileage')
plt.xlabel('Price ($)')
plt.ylabel('Mileage (MPG)')
plt.show()
```

10. Create a 3D scatter plot of the provided dataset with horsepower on the x-axis, weight on the y-axis, and miles per gallon on the z-axis.

```python
from mpl_toolkits.mplot3d import Axes3D

# Sample dataset for horsepower, weight, and MPG
car_data = {'Horsepower': [200, 300, 250, 150, 130],
            'Weight': [3500, 4000, 3800, 3000, 2800],
            'MPG': [20, 18, 22, 25, 30]}

df = pd.DataFrame(car_data)

# 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Horsepower'], df['Weight'], df['MPG'])
ax.set_xlabel('Horsepower')
ax.set_ylabel('Weight')
ax.set_zlabel('MPG')
plt.title('3D Scatter Plot: Horsepower vs Weight vs MPG')
plt.show()
```
